```python
income_map = {
    'Unknown' : 'N',
    'Less than $40K' : 'a',
    '$40K - $60K' : 'b',
    '$60K - $80K' : 'c',
    '$80K - $120K' : 'd',
    '$120K +’' : 'e'
}

df['newIncome'] = df['Income_Category'].map(income_map)
```

# 58

```python
def get_category(x):
    if x =='Unknown':
        return 'N'
    elif x =='Less than $40K':
        return 'a'
    elif x =='$40K - $60K':   
        return 'b'
    elif x =='$60K - $80K':    
        return 'c'
    elif x =='$80K - $120K':   
        return 'd'
    elif x =='$120K +' :     
        return 'e'
```

```python
df['newIncome'] = df['Income_Category'].apply(get_category)
```

# 67

```python
def fix_date(x):
    year = x.year
    month = x.month
    day = x.day
    
    if year >= 2061:
        year = year - 100
        
    return f'{year}-{month}-{day}'
```

```python
df['Yr_Mo_Dy'] = pd.to_datetime(df['Yr_Mo_Dy'].map(fix_date))
```
